```python
df = pd.read_excel('Sample_file.xlsx',sheet_name='Detail_Sheet_F001')

df = df.set_index(['User Name', 
                   'First Name',
                   'Last Name',
                   'User Lock Status']).stack().reset_index().rename(columns = {'level_4':'Category',
                                                                                0:'Score'})

df
```

```python
#fetch all the sheetnames
sheetnames = pd.ExcelFile('Sample_file.xlsx').sheet_names
sheetnames
```

```python
#Initialize and empty dataframe
df = pd.DataFrame()

for sheet in sheetnames:
    
    df1 = pd.read_excel('Sample_file.xlsx',sheet_name=sheet)
    
    df1 = df1.set_index(['User Name', 
                   'First Name',
                   'Last Name',
                   'User Lock Status']).stack().reset_index().rename(columns = {'level_4':'Category',
                                                                                0:'Score'})
    df = pd.concat([df,df1])

    
```

```python
df.reset_index(drop=True,inplace=True)
```
