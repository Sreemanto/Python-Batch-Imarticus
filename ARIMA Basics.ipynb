```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
temp = pd.read_csv('https://raw.githubusercontent.com/jbrownlee/Datasets/master/daily-min-temperatures.csv')
temp = temp[:365]
```

```python
from statsmodels.tsa.arima_model import ARIMA
```

```python
#p,d,q

#p is for AR (auto regressive part)
#d is for differencing
#q is for MA (Moving Average)

model = ARIMA(temp.Temp, order=(5,1,0)).fit() 
model.summary()
```

```python
plt.plot(temp.Temp)
plt.title('Original')
plt.show()

model.plot_predict()
plt.show()
```

```python
preds,error,conf = model.forecast(15)
preds
```

```python
model.plot_predict(365,379)
plt.show()
```
