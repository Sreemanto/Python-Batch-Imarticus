```python
from sklearn import datasets
import pandas as pd
```

```python
iris = datasets.load_iris()
df = pd.DataFrame(iris.data, columns = iris.feature_names)
df['Target'] = iris.get('target')
df
```

```python
x = df.drop(columns = 'Target')
y = df['Target']
```

```python
#before standardization
x.head(5)
```

```python
from sklearn.preprocessing import StandardScaler
```

```python
x = StandardScaler().fit_transform(x)
pd.DataFrame(x)
```

```python
from sklearn.decomposition import PCA
```

```python
#initializa PCA
pca = PCA(n_components= 3)
principal_components = pca.fit_transform(x)
principal_components
```

```python
#create this new dataset from PC's
df = pd.DataFrame(principal_components, columns = ['PC1','PC2','PC3'])
df['Target'] = iris.get('target')
df
```

```python
#Calculate the variance of each component

print("Variance of each PC's : ",pca.explained_variance_ratio_)
print("Total variation captured : ",sum(list(pca.explained_variance_ratio_))*100)
```

```python
#initializa PCA
pca = PCA(n_components= 2)
principal_components = pca.fit_transform(x)
#create this new dataset from PC's
df = pd.DataFrame(principal_components, columns = ['PC1','PC2'])
df['Target'] = iris.get('target')
df
```
