```python
req = requests.get('https://www.nytimes.com/interactive/2017/06/23/opinion/trumps-lies.html')
```

```python
from bs4 import BeautifulSoup
soup = BeautifulSoup(req.text,'html.parser')
```

```python
#<span class="short-desc">
records = soup.find_all('span',attrs={'class':'short-desc'})
```

```python
i = 5

DATE =        records[i].find('strong').text[:-1]
LIE =         records[i].contents[1][1:-2]
EXPLANATION = records[i].find('a').text[1:-1]
URL =         records[i].find('a')['href']

print('DATE : ',DATE)
print('LIE : ',LIE)
print('EXPLANATION : ',EXPLANATION)
print('URL : ',URL)
```

```python
#Collect all the records and create a DataFrame
results = []

for record in records:
    DATE =        record.find('strong').text[:-1]
    LIE =         record.contents[1][1:-2]
    EXPLANATION = record.find('a').text[1:-1]
    URL =         record.find('a')['href']
    
    results.append((DATE,LIE,EXPLANATION,URL))
```

```python
import pandas as pd
```

```python
#Converting the results in to DataFrame
df = pd.DataFrame(results,columns = ['DATE','LIE','EXPLANATION','URL'])
df.head(5)
```

```python
df.to_excel('Trump_Lies.xlsx',index=False)
```
