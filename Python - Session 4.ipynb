#### 1. How to extract all numbers between a given range

```python
a = np.random.randint(0,50,100)
a
```

```python
#>10 & <25
index = np.where((a>=10) & (a<=25))
```

```python
index
```

#### 3. How to replace a value over a praticular threshold to a constant

```python
# input = [2,3,10,11,12,5,6,7,4,5]
# output = [5,5,10,11,12,5,6,7,5,5]
```

```python
arr1 = np.random.randint(0,50,20)
arr1
```

```python
#Lower Threshold = 10
# Upper Threshold = 40
arr1 = np.clip(arr1, a_min=10, a_max=40)
arr1
```
