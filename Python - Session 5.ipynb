#### 3. How to replace a value over a praticular threshold to a constant

```python
# input = [2,3,10,11,12,5,6,7,4,5]
# output = [5,5,10,11,12,5,6,7,5,5]
```

```python
arr1 = np.random.randint(0,50,20)
arr1
```

```python
#Lower Threshold = 10
# Upper Threshold = 40
arr1 = np.clip(arr1, a_min=10, a_max=40)
arr1
```

```python
import matplotlib.pyplot as plt # the oldest
import seaborn as sns           # bulit on top of matplotlib. Its a beautified version of matplotlib
```

#### Line plots

```python
yield_apples = [0.895, 0.91 ,0.919, 0.926, 0.929, 0.931]
```

```python
plt.plot(yield_apples)
plt.show()
```

```python
yield_apples = [0.895, 0.91 ,0.919, 0.926, 0.929, 0.931]
years = [2015,2016,2017,2018,2019,2020]
```

```python
plt.plot(years,yield_apples)
plt.title('Yearwise Yield of apple',size =20)
plt.xlabel('Years')
plt.ylabel('Yield (tons per hectare)')
plt.show()
```

```python
#plotting multiple lines
yield_apples = [0.895, 0.91 ,0.919, 0.926, 0.929, 0.931]
yield_oranges = [0.962, 0.941, 0.930, 0.923, 0.908, 0.898]

years = [2015,2016,2017,2018,2019,2020]

plt.plot(years,yield_apples, marker='o', c='red',ls='dashed',lw=4)
plt.plot(years,yield_oranges, marker ='*',c='green',ls='dotted',lw=2)

plt.title('Yearwise Yield of fruits',size =20)
plt.xlabel('Years')
plt.ylabel('Yield (tons per hectare)')
plt.legend(['Apples','Oranges'])
plt.show()
```
